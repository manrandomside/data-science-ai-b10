**0. Import module yang diperlukan**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

**1. Membaca data**


> Silakan gunakan teknik membaca data teratas dan terbawah (cth: head(), tail()) yang tersedia pada pandas, atau teknik lainnya seperti describe()



```python
df = pd.read_csv('Housing.csv')
df.head()
```

Insight :

> Dataset berisi data harga rumah dengan beberapa fitur seperti area (luas), jumlah kamar tidur, kamar mandi, lantai, dan beberapa fasilitas seperti AC, parkir, dll. Tipe datanya campuran antara numerik (price, area, bedrooms) dan kategorik (mainroad, guestroom, basement, dll yang isinya yes/no), jadi nanti perlu transformasi ke angka supaya bisa dianalisis lebih lanjut.


**2. Cek Missing values dan duplicate pada data**
> Cek apakah ada missing values / duplicate pada data, jika ada perbaiki, jika tidak ada, berikan insight/penjelasan.

```python
print("Shape:", df.shape)
print()
print("Missing values per kolom:")
print(df.isnull().sum())
```

```python
print("Jumlah duplicate:", df.duplicated().sum())
print()
print("Tipe data:")
print(df.dtypes)
```

Insight :

> Setelah dicek, dataset bersih, tidak ada missing values di kolom manapun dan tidak ada baris yang duplikat. Dari segi tipe data, kolom numerik seperti price, area, bedrooms, bathrooms, stories, dan parking sudah dalam bentuk integer. Sementara kolom seperti mainroad, guestroom, basement, hotwaterheating, airconditioning, prefarea, dan furnishingstatus masih object (string), nanti perlu di-transform ke numerik untuk keperluan analisis dan modeling.


**3. Hapus data jika ada duplicate dan isi missing values jika ada**

```python
df = df.drop_duplicates().reset_index(drop=True)
print("Shape setelah hapus duplicate:", df.shape)
```

```python
print("Cek ulang missing values:")
print(df.isnull().sum().sum(), "missing values total")
```

Insight :

> Karena dataset ini sudah bersih dari awal, hapus duplicate tidak mengubah jumlah baris dan tidak ada missing values yang perlu diisi. Tapi langkah ini tetap dilakukan sebagai best practice supaya kalau dataset di-update di masa depan, kode-nya tetap aman dipakai.


**4. Transformasi format**

> Cek dataset, apakah ada data yang bisa ditransformasi? (cth : yes or no menjadi 1 dan 0)

```python
kolom_yn = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

for kol in kolom_yn:
    df[kol] = df[kol].map({'yes': 1, 'no': 0})

df['furnishingstatus'] = df['furnishingstatus'].map({
    'unfurnished': 0,
    'semi-furnished': 1,
    'furnished': 2
})

df.head()
```

Insight :

> Semua kolom kategorik berhasil ditransformasi ke numerik. Untuk kolom yes/no diubah jadi 1/0, sedangkan furnishingstatus yang punya 3 kategori diubah jadi 0 (unfurnished), 1 (semi-furnished), dan 2 (furnished) karena ada urutan tingkatannya. Sekarang semua kolom sudah berbentuk angka dan siap untuk analisis korelasi serta modeling.


**5. Laksanakan EDA (Exploratory Data Analysis)**

> Pada tahap ini, lakukan visualisasi data, dan eksplorasi serta analisis dataset yang diberikan sesuai EDA, lalu berikan setidaknya 3 insight dari dataset tersebut. Silakan tambah cell sebanyak - banyak yang diperlukan untuk keperluan EDA atau visualisasi data.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(df['price'], bins=30, color='steelblue', edgecolor='black', alpha=0.8)
axes[0].set_title('Distribusi Harga Rumah')
axes[0].set_xlabel('Harga')
axes[0].set_ylabel('Jumlah Rumah')

axes[1].hist(df['area'], bins=30, color='seagreen', edgecolor='black', alpha=0.8)
axes[1].set_title('Distribusi Luas Area')
axes[1].set_xlabel('Area')
axes[1].set_ylabel('Jumlah Rumah')

plt.tight_layout()
plt.show()

print("Statistik harga:")
print(df['price'].describe())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(data=df, x='bedrooms', y='price', ax=axes[0], palette='Blues')
axes[0].set_title('Harga Rumah Berdasarkan Jumlah Kamar Tidur')

sns.boxplot(data=df, x='stories', y='price', ax=axes[1], palette='Greens')
axes[1].set_title('Harga Rumah Berdasarkan Jumlah Lantai')

plt.tight_layout()
plt.show()
```

Insight :

> Dari analisis EDA ada 3 temuan utama:
>
> 1. Distribusi harga rumah cenderung skewed ke kanan, artinya mayoritas rumah harganya di kisaran rendah-menengah, tapi ada beberapa rumah dengan harga jauh lebih tinggi (outlier). Hal serupa juga terjadi di luas area.
>
> 2. Jumlah kamar tidur ada hubungan dengan harga, makin banyak kamar makin tinggi harganya, tapi tidak linear. Rumah dengan 4 kamar punya range harga yang paling lebar, ada yang murah ada yang mahal.
>
> 3. Jumlah lantai juga berpengaruh ke harga, rumah 4 lantai cenderung paling mahal, tapi rumah 1 lantai pun bisa mahal kalau fasilitasnya lengkap atau lokasinya bagus.


**6. Correlation Matrix**

> Buat correlation matrix, transformasi dulu seluruh data ke numerik agar tidak error saat masuk ke correlation matrix. Lalu berikan insight, apa yang kamu temukan dari correlation matrix dataset tersebut?

```python
korelasi = df.corr()

plt.figure(figsize=(11, 8))
sns.heatmap(korelasi, annot=True, fmt='.2f', cmap='RdBu_r', center=0, linewidths=0.5, square=True)
plt.title('Heatmap Korelasi Antar Variabel')
plt.tight_layout()
plt.show()

print("Korelasi terhadap harga:")
print(korelasi['price'].sort_values(ascending=False))
```

Insight :

> Dari heatmap korelasi terlihat bahwa variabel yang paling berpengaruh terhadap harga rumah adalah area (luas tanah) dengan korelasi positif paling tinggi. Diikuti bathrooms, airconditioning, dan stories. Yang menarik, jumlah bedrooms ternyata korelasinya tidak setinggi yang dikira, ini menunjukkan bahwa jumlah kamar tidur saja tidak otomatis bikin rumah mahal, tapi luas dan fasilitas pendukung lebih menentukan. Kolom seperti mainroad dan prefarea juga punya korelasi positif yang menunjukkan lokasi strategis ikut menambah nilai jual rumah.


**7. Feature Engineering**



> Laksanakan Feature Engineering, buat variabel baru berdasarkan variabel yang sudah ada. Analisis hasil dari variabel baru tersebut.



```python
df['price_per_area'] = df['price'] / df['area']
df['total_rooms'] = df['bedrooms'] + df['bathrooms']

df[['price', 'area', 'price_per_area', 'bedrooms', 'bathrooms', 'total_rooms']].head()
```

Insight :

> Dibuat 2 variabel baru:
>
> 1. price_per_area = harga per satuan luas. Variabel ini lebih adil untuk membandingkan rumah karena rumah besar otomatis mahal, tapi belum tentu mahal per meternya. Dengan variabel ini bisa dilihat rumah mana yang harganya premium relatif terhadap luasnya.
>
> 2. total_rooms = bedrooms + bathrooms. Variabel ini menggabungkan total ruangan utama yang biasanya jadi pertimbangan pembeli. Rumah dengan total_rooms tinggi biasanya cocok untuk keluarga besar.
>
> Variabel-variabel turunan seperti ini berguna untuk modeling karena bisa menangkap informasi yang tidak terlihat di variabel asli.
